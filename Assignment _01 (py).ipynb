```python
#problem-18
#Solution...


score = int(input("Enter student score (0-100): "))

if 90 <= score <= 100:
    grade = 'A+'
elif 80 <= score <= 89:
    grade = 'A'
elif 60 <= score <= 79:
    grade = 'B'
elif 40 <= score <= 59:
    grade = 'C'
else:
    grade = 'F'

if score >= 40:
    status = 'Pass'
else:
    status = 'Fail'

print(f"Grade: {grade} | Status: {status}")
```
